### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define data
x = np.arange(1, 6)
y = x ** 2
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]

# Create figure and subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Subplot 1: Line Plot
axes[0, 0].plot(x, y, marker='o', linestyle='-', color='b')
axes[0, 0].set_title("Line Plot")
axes[0, 0].set_xlabel("X")
axes[0, 0].set_ylabel("Y")

# Subplot 2: Bar Chart
axes[0, 1].bar(categories, values, color=['blue', 'green', 'red', 'purple'])
axes[0, 1].set_title("Bar Chart")
axes[0, 1].set_xlabel("Categories")
axes[0, 1].set_ylabel("Values")

# Subplot 3: Scatter Plot
axes[1, 0].scatter(x, y, color='orange', marker='o')
axes[1, 0].set_title("Scatter Plot")
axes[1, 0].set_xlabel("X")
axes[1, 0].set_ylabel("Y")

# Subplot 4: Pie Chart
axes[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', colors=['blue', 'green', 'red', 'orange'])
axes[1, 1].set_title("Pie Chart")

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
